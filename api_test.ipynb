```python
params = {"name": "Ростовская область, Ростов-на-Дону"}
```

```python
import requests

# Определение URL и параметров запроса
url = "https://6219-5-45-70-209.ngrok-free.app/factory_api/flows/optimize"


# Определение заголовков запроса
headers = {"accept": "application/json"}

try:
    # Выполнение GET-запроса
    response = requests.get(url, params=params, headers=headers)

    # Проверка успешности запроса
    response.raise_for_status()  # Возбудит исключение для HTTP ошибок

    # Получение данных в формате JSON
    data = response.json()

    # Вывод результата
    print("Статус код:", response.status_code)
    print("Ответ сервера:", data)

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP ошибка произошла: {http_err}")
except requests.exceptions.ConnectionError as conn_err:
    print(f"Ошибка соединения: {conn_err}")
except requests.exceptions.Timeout as timeout_err:
    print(f"Истекло время ожидания запроса: {timeout_err}")
except requests.exceptions.RequestException as req_err:
    print(f"Произошла ошибка при запросе: {req_err}")
except ValueError:
    print("Ошибка декодирования JSON.")
```
