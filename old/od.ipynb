```python
import itertools
import numpy as np
import networkx as nx


def jaccard_coefficient(G):
    jaccard_scores = {}
    for u, v in itertools.combinations(G.nodes(), 2):
        # if G.has_edge(u, v):
        #     continue
        neighbors_u = set(G.neighbors(u))
        neighbors_v = set(G.neighbors(v))
        intersection_size = len(neighbors_u.intersection(neighbors_v))
        union_size = len(neighbors_u.union(neighbors_v))
        jaccard_scores[(u, v)] = intersection_size / union_size if union_size > 0 else 0
    return jaccard_scores


def adamic_adar_index(G):
    aa_scores = {}
    for u, v in itertools.combinations(G.nodes(), 2):
        # if G.has_edge(u, v):
        #     continue
        common_neighbors = set(G.neighbors(u)).intersection(set(G.neighbors(v)))
        aa_score = sum(
            1 / np.log(len(set(G.neighbors(x))))
            for x in common_neighbors
            if len(set(G.neighbors(x))) > 0
        )
        aa_scores[(u, v)] = aa_score
    return aa_scores


# """
# NOT IN USAGE i
#              v
# """


# def preferential_attachment(G):
#     pa_scores = {}
#     for u, v in itertools.combinations(G.nodes(), 2):
#         # if G.has_edge(u, v):
#         #     continue
#         pa_scores[(u, v)] = len(list(G.neighbors(u))) * len(list(G.neighbors(v)))
#     return pa_scores


def common_neighbors(G):
    cn_scores = {}
    for u, v in itertools.combinations(G.nodes(), 2):
        # if G.has_edge(u, v):
        #     continue
        cn_scores[(u, v)] = len(set(G.neighbors(u)).intersection(set(G.neighbors(v))))
    return cn_scores


# Calculate scores
jaccard_scores = jaccard_coefficient(g)
adamic_adar_scores = adamic_adar_index(g)
# pa_scores = preferential_attachment(g)
cn_scores = common_neighbors(g)

# Display results
print("Jaccard Coefficients:", jaccard_scores)
print("Adamic-Adar Index Scores:", adamic_adar_scores)
# print("Preferential Attachment Scores:", pa_scores)
print("Common Neighbors Scores:", cn_scores)
```

```python
# Create a DataFrame from the scores
score_data = {
    "Node1": [],
    "Node2": [],
    "Edge Weight": [],
    "Jaccard Coefficient": [],
    "Adamic-Adar Index": [],
    "Preferential Attachment": [],
    "Common Neighbors": [],
}

# Fill the DataFrame with scores
for u, v in itertools.combinations(g.nodes(), 2):
    score_data["Node1"].append(u)
    score_data["Node2"].append(v)
    score_data["Edge Weight"].append(
        g[u][v]["weight"] if g.has_edge(u, v) else 0
    )  # Get edge weight
    score_data["Jaccard Coefficient"].append(jaccard_scores.get((u, v), 0))
    score_data["Adamic-Adar Index"].append(adamic_adar_scores.get((u, v), 0))
    # score_data["SIM"].append(sim.get((u, v), 0))
    # score_data["Preferential Attachment"].append(pa_scores.get((u, v), 0))
    score_data["Common Neighbors"].append(cn_scores.get((u, v), 0))

# Create the DataFrame
scores_df = pd.DataFrame(score_data)

# Display the DataFrame
display(scores_df)
```

```python
import numpy as np
import pandas as pd

scores_df = pd.read_parquet("scores_df.parquet")
scores_df.loc[scores_df["Edge Weight"] < 10, "Edge Weight"] = np.nan
scores_df.loc[scores_df["Edge Weight"] > 50, "Edge Weight"] = np.nan
```

```python
scores_df.sample()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

scaler_x = MinMaxScaler()

subset_existing = scores_df[scores_df["Edge Weight"].notna()]
# Define features and target variable
X = subset_existing.drop(
    columns=["Node1", "Node2", "Edge Weight", "Preferential Attachment"]
).reset_index(drop=True)


X = np.log(X + 1)
X = scaler_x.fit_transform(X)

y = subset_existing["Edge Weight"].reset_index(drop=True)
```

```python
from smogn import smoter  # или можно использовать imbalanced_learn для SMOTER

# Создаем сбалансированный датасет для регрессии
X_resampled = smoter(
    data=pd.concat([pd.DataFrame(X), y.to_frame()], axis=1).dropna(),
    y="Edge Weight",
    k=3,
    pert=0.1,
    rel_thres=0.04,  # порог для определения редких значений
    rel_method="auto",  # автоматическое определение редких значений
    samp_method="extreme",
)

# Разделяем признаки и целевую переменную
y_resampled = X_resampled["Edge Weight"]
X_resampled = X_resampled.drop("Edge Weight", axis=1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42
)

# X_train, X_test, y_train, y_test = train_test_split(
#     X, y, test_size=0.3, random_state=42
# )
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error

model = CatBoostRegressor(
    loss_function="MAE",  # или 'Tweedie' с параметром variance_power
    eval_metric="MAPE",
    iterations=500,
    learning_rate=0.05,
    depth=12,
    random_seed=42,
    early_stopping_rounds=50,
)

model.fit(
    X_train,
    y_train,
    # sample_weight=sample_weights,
    eval_set=(
        pd.DataFrame(X).sample(20, random_state=42),
        y.sample(20, random_state=42),
    ),
    early_stopping_rounds=20,
    verbose=20,
)
```
